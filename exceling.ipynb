# Tratamento

## Tratando Campo de Endereço, separando em várias colunas

```python
df_empresas= pd.read_csv('empresas.csv')
```

```python
df_empresas.head()
```

```python
df_empresas = df_empresas.drop('Unnamed: 0', axis=1)
```

```python
pd.set_option('display.max_colwidth', None)  # Permite exibir o texto completo de qualquer célula
print(df_empresas['endereco'].head())
```

```python
#df_empresas['Logradouro'] = df_empresas['endereco'].str.extract(r'Logradouro:(.*?), (\d+)')[0]
#df_empresas['Logradouro'] = df_empresas['endereco'].str.extract(r'Logradouro:(.*?)(?:Bairro:|Município/UF:|CEP:|$)')[0]
df_empresas['Logradouro'] = df_empresas['endereco'].str.extract(r'Logradouro:(.*?)(?:, \d+|Bairro:|Município/UF:|CEP:|$)')[0]
df_empresas['Número'] = df_empresas['endereco'].str.extract(r'Logradouro:(.*?), (\d+)')[1]
df_empresas['Bairro'] = df_empresas['endereco'].str.extract(r'Bairro:(.*?)(?:Município|Município/UF|CEP)')
df_empresas['Município'] = df_empresas['endereco'].str.extract(r'Município/UF:(.*?), ([A-Z]{2})')[0]
df_empresas['Estado'] = df_empresas['endereco'].str.extract(r'Município/UF:(.*?), ([A-Z]{2})')[1]
df_empresas['CEP'] = df_empresas['endereco'].str.extract(r'CEP:(\d{2}\.\d{3}-\d{3})')

df_empresas = df_empresas.drop('endereco', axis=1)
```
